```python
import pandas as pd
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report
from statistics import mean
```

```python
diseases_test = pd.read_csv('diseases_table_test.csv')
diseases_results = pd.read_csv('diseases_table_results.csv')
drugs_test = pd.read_csv('drugs_table_test.csv')
drugs_results = pd.read_csv('drugs_table_results.csv')
```

```python
scores_file = open('scores_file.txt', 'w')
print('CLASSIFICATION SCORES', file=scores_file)
print('\n', file=scores_file)

test_drugs = list(drugs_test['standart_names'])
results_drugs = list(drugs_results['standart_names'])
union_drugs = list(set(test_drugs + results_drugs))
intersection_drugs = []
for drug in union_drugs:
    if drug in test_drugs and drug in results_drugs:
        intersection_drugs.append(drug)
print('Количество препаратов в объединении: ', len(union_drugs))
print('Количество препаратов в пересечении: ', len(intersection_drugs))
print('Количество препаратов в original:', len(test_drugs))
print('Количество препаратов в results:', len(results_drugs))
print('The ratio of the number of drugs in the intersection to the number of drugs in the original: ', round(len(intersection_drugs) / len(test_drugs), 2), file=scores_file)
print('The ratio of the number of drugs in the intersection to the number of drugs in the results: ', round(len(intersection_drugs) / len(results_drugs), 2), file=scores_file)
```

```python
test_diseases = list(diseases_test['disease'])
results_diseases = list(diseases_results['disease'])
union_diseases = list(set(test_diseases + results_diseases))
intersection_diseases = []
for disease in union_diseases:
    if disease in test_diseases and disease in results_diseases :
        intersection_diseases.append(disease)
print('Количество симптомов в объединении: ', len(union_diseases))
print('Количество симптомов в пересечении: ', len(intersection_diseases))
print('Количество симптомов в original:', len(test_diseases))
print('Количество симптомов в results:', len(results_diseases))
print('Отношение количества симптомов в пересечении к количеству симптомов в original: ', round(len(intersection_diseases) / len(test_diseases), 2))
print('Отношение количества симптомов в пересечении к количеству симптомов в results: ', round(len(intersection_diseases) / len(results_diseases), 2))
```

```python
test_exclusive_drugs, results_exclusive_drugs = [], []
for drug in test_drugs:
    if drug not in intersection_drugs:
        test_exclusive_drugs.append(drug)
for drug in results_drugs:
    if drug not in intersection_drugs:
        results_exclusive_drugs.append(drug)
print(test_exclusive_drugs)
print(len(test_exclusive_drugs))
print(results_exclusive_drugs)
print(len(results_exclusive_drugs))
```

```python
test_exclusive_diseases, results_exclusive_diseases = [], []
for disease in test_diseases:
    if disease not in intersection_diseases:
        test_exclusive_diseases.append(disease)
for disease in results_diseases:
    if disease not in intersection_diseases:
        results_exclusive_diseases.append(disease)
print(test_exclusive_diseases)
print(len(test_exclusive_diseases))
print(results_exclusive_diseases)
print(len(results_exclusive_diseases))
```

```python
drugs_test1 = drugs_test.set_index('standart_names')
drugs_test2 = drugs_test1.drop(test_exclusive_drugs, axis=0)
drugs_test2 = drugs_test2.sort_values(by=['standart_names'], ascending=True)
drugs_test2.head(20)
```

```python
drugs_results1 = drugs_results.set_index('standart_names')
drugs_results2 = drugs_results1.drop(results_exclusive_drugs, axis=0)
drugs_results2 = drugs_results2.sort_values(by=['standart_names'], ascending=True)
drugs_results2.head(20)
```

```python
diseases_test1 = diseases_test.set_index('disease')
diseases_test2 = diseases_test1.drop(test_exclusive_diseases, axis=0)
diseases_test2 = diseases_test2.sort_values(by=['disease'], ascending=True)
diseases_test2.head(20)
```

```python
diseases_results1 = diseases_results.set_index('disease')
diseases_results2 = diseases_results1.drop(results_exclusive_diseases, axis=0)
diseases_results2 = diseases_results2.sort_values(by=['disease'], ascending=True)
diseases_results2.head(20)
```

```python
test_raiting = list(drugs_test2['Raiting'])
results_raiting = list(drugs_results2['Raiting'])
test_adr_str = list(drugs_test2['ADR'])
results_adr_str = list(drugs_results2['ADR'])
test_good_drugs = list(diseases_test2['drug'])
results_good_drugs = list(diseases_results2['drug'])
```

```python
print('\n', file=scores_file)
print('Classification report for drugs raiting (scores for 5 classes): ', file=scores_file)
print(classification_report(test_raiting, results_raiting), file=scores_file)
```

```python
test_adr = []
results_adr = []
for i in range(len(test_adr_str)):
    temp_test = test_adr_str[i].replace('[', '').replace("'", '').replace(']', '')
    temp_test_lst = temp_test.split(',')
    test_adr.append(temp_test_lst)
    temp_results = results_adr_str[i].replace('[', '').replace("'", '').replace(']', '')
    temp_results_lst = temp_results.split(',')
    results_adr.append(temp_results_lst)
print(test_adr)
print(results_adr)
```

```python
union_good_drugs = []
test_good_drugs_lst = []
results_good_drugs_lst = []
for i in range(len(test_good_drugs)):
    temp_union, temp_union1 = [], []
    temp_union_test, temp_union_results = [], []
    temp_union_test1, temp_union_results1 = [], []
    for elem in test_good_drugs[i].split(','):
        temp_union.append(elem)
        temp_union_test.append(elem)
    for elem in results_good_drugs[i].split(','):
        temp_union.append(elem)
        temp_union_results.append(elem)
    for elem in temp_union:
        if elem[0] == ' ':
            elem = elem.replace(' ', '', 1)
        temp_union1.append(elem)
    for elem in temp_union_test:
        if elem[0] == ' ':
            elem = elem.replace(' ', '', 1)
        temp_union_test1.append(elem)
    for elem in temp_union_results:
        if elem[0] == ' ':
            elem = elem.replace(' ', '', 1)
        temp_union_results1.append(elem)
    temp_union1 = list(set(temp_union1))
    union_good_drugs.append(temp_union1)
    test_good_drugs_lst.append(temp_union_test1)
    results_good_drugs_lst.append(temp_union_results1)
print(union_good_drugs)
print(test_good_drugs_lst)
print(results_good_drugs_lst)
```

```python
union_adr = []
test_adr_lst = []
results_adr_lst = []
for i in range(len(test_adr)):
    temp_union, temp_union1 = [], []
    temp_union_test, temp_union_results = [], []
    temp_union_test1, temp_union_results1 = [], []
    for elem in test_adr[i]:
        temp_union.append(elem)
        temp_union_test.append(elem)
    for elem in results_adr[i]:
        temp_union.append(elem)
        temp_union_results.append(elem)
    for elem in temp_union:
        if elem != '':
            if elem[0] == ' ':
                elem = elem.replace(' ', '', 1)
            temp_union1.append(elem)
    for elem in temp_union_test:
        if elem != '':
            if elem[0] == ' ':
                elem = elem.replace(' ', '', 1)
            temp_union_test1.append(elem)
    for elem in temp_union_results:
        if elem != '':
            if elem[0] == ' ':
                elem = elem.replace(' ', '', 1)
            temp_union_results1.append(elem)
    temp_union1 = list(set(temp_union1))
    union_adr.append(temp_union1)
    test_adr_lst.append(temp_union_test1)
    results_adr_lst.append(temp_union_results1)
print(union_adr)
print(test_adr_lst)
print(results_adr_lst)
print(len(union_adr))
print(len(test_adr_lst))
print(len(results_adr_lst))
```

```python
drugs_precisions = []
drugs_recalls = []
drugs_f1 = []
for i in range(len(test_good_drugs_encoded)):
    drugs_precisions.append(round(precision_score(test_good_drugs_encoded[i], results_good_drugs_encoded[i]), 3))
    drugs_recalls.append(round(recall_score(test_good_drugs_encoded[i], results_good_drugs_encoded[i]), 3))
    drugs_f1.append(round(f1_score(test_good_drugs_encoded[i], results_good_drugs_encoded[i]), 3))
print(drugs_precisions)
print(drugs_recalls)
print(drugs_f1)
print('\n', file=scores_file)
print('Precision score for drugs: ', round(mean(drugs_precisions), 3), file=scores_file)
print('Recall score for drugs: ', round(mean(drugs_recalls), 3), file=scores_file)
print('f1 score for drugs: ', round(mean(drugs_f1), 3), file=scores_file)


adr_precisions = []
adr_recalls = []
adr_f1 = []
for i in range(len(test_adr_encoded)):
    adr_precisions.append(round(precision_score(test_adr_encoded[i], results_adr_encoded[i]), 3))
    adr_recalls.append(round(recall_score(test_adr_encoded[i], results_adr_encoded[i]), 3))
    adr_f1.append(round(f1_score(test_adr_encoded[i], results_adr_encoded[i]), 3))
print(adr_precisions)
print(adr_recalls)
print(adr_f1)
print('Precision score for ADR: ', round(mean(adr_precisions), 3), file=scores_file)
print('Recall score for ADR: ', round(mean(adr_recalls), 3), file=scores_file)
print('f1 score for ADR: ', round(mean(adr_f1), 3), file=scores_file)
```
